## Generating Path Graphs
- Pk ~ Path of length k

```python
def genPath(k):
    #ret = [[0 for _ in range(k+1)] for _ in range(k+1)]
    ret = []
    for i in range(k):
        row = [0 for _ in range(k)]
        if i > 0:
            row[i - 1] += 1
        if i < k-1:
            row[i + 1] += 1
        ret.append(row)
    return ret

cycle = 6
path = 10

G4_1 = nx.from_numpy_matrix(np.array(genPath(path)))
G4_2 = nx.path_graph(path)

print(nx.is_isomorphic(G4_1,G4_2))

A = nx.adjacency_matrix(G4_1).todense()
#for i in range(0,21,2):
#    print('i =', i)
#    print(A**i)
#    print('')
#    print('=================')

b = A**cycle
print(np.trace(b))
print(b)

homs = 0



for i in range(1,(path)//2 + 1):
    if path - i < 0:
        d = scipy.special.binom(cycle, cycle/2)
    else:
        d = scipy.special.binom(cycle, cycle/2) - scipy.special.binom(cycle, cycle/2 - i)

    homs += 2*d
homs = int(homs)

print(homs)
```
